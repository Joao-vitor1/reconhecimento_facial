```python
import matplotlib.pyplot as plt
from PIL import Image
import torchvision.transforms as T
from torchvision.datasets import ImageFolder
```

```python
transform_val = T.Compose(
    [
        T.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.6),
       # T.CenterCrop(140), #retirando bordas vermelhas em imagens 144x144
        T.RandomHorizontalFlip(p = 0.5), #rotação horizontal com probabilidade de 50%
        T.RandomRotation(15, fill = (255)), #rotação de até 15 graus na imagem
        T.ToTensor()

    ]
)
```

```python
figure = plt.figure(figsize=(5,5))
#insira o nome da sua foto que deseja alterar
foto = Image.open('joao.jpg')

for i in range(4):
  foto2 = transform_val(foto)

  figure.add_subplot(2,2,i+1)
  plt.axis('off')
  plt.imshow(foto2.permute(1,2,0))
plt.show()
```
